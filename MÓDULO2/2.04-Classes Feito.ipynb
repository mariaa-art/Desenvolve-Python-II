#### Q1.
Vamos criar aos pouquinhos uma classe Evento que registra eventos de um calend√°rio. Pra come√ßar, nessa quest√£o voc√™ deve:

* Criar uma classe chamada Evento
* Adicionar um atributo de classe `total_eventos` que ser√° usado para contar o n√∫mero total de eventos (inst√¢ncias de classe) criadas.
* Implemente o m√©todo construtor que deve receber os par√¢metros `titulo` (`string`), `data_hora` (`datetime`), `descri√ß√£o` (`string`). Crie e inicialize um atributo de inst√¢ncia para cada um dos par√¢metros citados.
* O construtor tamb√©m deve inicializar um atributo de inst√¢ncia `is_concluido = False`. Esse atributo ser√° utilizado mais √† frente. 
* No construtor voc√™ tamb√©m deve incrementar o atributo de classe `total_eventos` em 1 a cada nova inst√¢ncia criada.

Teste o seu c√≥digo criando duas diferentes inst√¢ncias de evento (preencha os atributos como quiser) e imprimindo todos os seus atributos (de classe e de inst√¢ncia).

> Consulte [este tutorial biblioteca `datetime`](https://www.w3schools.com/python/python_datetime.asp) caso n√£o esteja familiarizado.

```python
import datetime

class Evento:
    total_eventos = 0 # Contador global de eventos

    def __init__(self, titulo, data_hora, descricao):
        # Guarda as informacoes de cada evento novo
        self.titulo = titulo
        self.data_hora = data_hora
        self.descricao = descricao
        self.is_concluido = False
        
        # Soma +1 no total de eventos da classe
        Evento.total_eventos += 1

# --- testando o codigo ---

ev1 = Evento(
    "Reuniao de Alinhamento", 
    datetime.datetime(2025, 12, 20, 10, 30), 
    "Conversa sobre metas"
)

ev2 = Evento(
    "Workshop de Python", 
    datetime.datetime(2026, 1, 15, 14, 0), 
    "Pratica de classes e objetos"
)

# Verificando os dados
print(f"Evento 1: {ev1.titulo}, {ev1.data_hora}, {ev1.is_concluido}")
print(f"Evento 2: {ev2.titulo}, {ev2.data_hora}, {ev2.is_concluido}")
print(f"Total no sistema: {Evento.total_eventos}")
```

```python
import datetime

class Evento:
    total_eventos = 0 

    def __init__(self, titulo, data_hora, descricao):
        self.titulo = titulo
        self.data_hora = data_hora
        self.descricao = descricao
        self.is_concluido = False
        Evento.total_eventos += 1

    # Atualiza para True se a data ja passou
    def isConcluido(self):
        if self.data_hora < datetime.datetime.now():
            self.is_concluido = True
        return self.is_concluido

    # Retorna o contador da classe
    @classmethod
    def num_eventos(cls):
        return cls.total_eventos

    # Valida se os tipos de entrada estao corretos
    @staticmethod
    def valida_evento(nome, data_hora, descricao):
        val_nome = isinstance(nome, str)
        val_data = isinstance(data_hora, datetime.datetime)
        val_desc = isinstance(descricao, str)
        return val_nome and val_data and val_desc

# --- testando a Q2 ---

# Instancia com data passada para testar conclusao
ev_passado = Evento("Ontem", datetime.datetime(2023, 1, 1), "Evento que ja foi")
ev_passado.isConcluido()
print(f"Evento concluido? {ev_passado.is_concluido}")

# Testando metodo de classe
print(f"Total de eventos via metodo: {Evento.num_eventos()}")

# Testando metodo estatico
print(f"Validacao correta: {Evento.valida_evento('Teste', datetime.datetime.now(), 'Desc')}")
print(f"Validacao errada: {Evento.valida_evento(123, 'nao eh data', None)}")
```

#### Q3.

Vamos incluir m√©todos m√°gicos! ‚ú®ü™ÑüîÆ

Redefina a classe incluindo":
* M√©todo `__str__` que imprime os atributos do evento na forma `"Evento: titulo, Data: data_hora, Descri√ß√£o: descricao, Concluido: is_concluido"`.
* Implemente os m√©todos de compara√ß√£o `__eq__`, `__ne__`, `__lt__`, `__le__`, `__gt__` e `__ge__` para comparar eventos baseados no atributo `data_hora`. Esses m√©todos devem comparar duas inst√¢ncias de Evento e retornar os resultados apropriados (`True` ou `False`).

Para testar, crie duas inst√¢ncias de Evento com datas diferentes. Imprima as inst√¢ncias com a fun√ß√£o `print()` e apresente o resultado das compara√ß√µes entre eventos (`==`, `!=`, `<`, `<=`, `>`, `>=`).

```python
import datetime

class Evento:
    total_eventos = 0 

    def __init__(self, titulo, data_hora, descricao):
        self.titulo = titulo
        self.data_hora = data_hora
        self.descricao = descricao
        self.is_concluido = False
        Evento.total_eventos += 1

    def isConcluido(self):
        if self.data_hora < datetime.datetime.now():
            self.is_concluido = True
        return self.is_concluido

    @classmethod
    def num_eventos(cls):
        return cls.total_eventos

    @staticmethod
    def valida_evento(nome, data_hora, descricao):
        val_nome = isinstance(nome, str)
        val_data = isinstance(data_hora, datetime.datetime)
        val_desc = isinstance(descricao, str)
        return val_nome and val_data and val_desc

    # Metodo para imprimir os dados do evento de forma organizada
    def __str__(self):
        return f"Evento: {self.titulo}, Data: {self.data_hora}, Descricao: {self.descricao}, Concluido: {self.is_concluido}"

    # Metodos de comparacao baseados na data e hora
    def __eq__(self, outro):
        return self.data_hora == outro.data_hora

    def __ne__(self, outro):
        return self.data_hora != outro.data_hora

    def __lt__(self, outro):
        return self.data_hora < outro.data_hora

    def __le__(self, outro):
        return self.data_hora <= outro.data_hora

    def __gt__(self, outro):
        return self.data_hora > outro.data_hora

    def __ge__(self, outro):
        return self.data_hora >= outro.data_hora

# --- testando a Q3 ---

# Criando dois eventos com datas diferentes
ev_a = Evento("Festa", datetime.datetime(2026, 5, 20, 22, 0), "Comemoracao")
ev_b = Evento("Trabalho", datetime.datetime(2026, 6, 10, 9, 0), "Reuniao")

# Testando o __str__
print(ev_a)
print(ev_b)

# Testando as comparacoes
print(f"ev_a == ev_b: {ev_a == ev_b}")
print(f"ev_a != ev_b: {ev_a != ev_b}")
print(f"ev_a < ev_b:  {ev_a < ev_b}")
print(f"ev_a <= ev_b: {ev_a <= ev_b}")
print(f"ev_a > ev_b:  {ev_a > ev_b}")
print(f"ev_a >= ev_b: {ev_a >= ev_b}")
```
